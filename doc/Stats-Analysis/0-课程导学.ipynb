## 3. 课程基础


### 3.1 数学基础

1. 具备高等数学或微积分基础知识，能看懂模型的公式。
2. 线性代数基础，包括：
   - 向量计算
   - 矩阵计算
   - 矩阵特征值分解与奇异值分解
   - 范数和行列式
   - 线性方程组求解
   - 至少能在指导下看得懂线性代数和矩阵计算的内在机制和过程。
3. 概率论与数理统计基础，了解：
   - 基本的概率知识
   - 各种概率分布
   - 统计推断的基础知识（点估计、区间估计、假设检验等）
   - 回归分析
   - 方差分析
   - 基本的应用多元统计知识（主成分分析、因子分析、判别分析、聚类分析、典型相关分析等）
   - 非参数统计基础知识

### 3.2 Python

1. 掌握Python基本语法。
2. 掌握Numpy和Scipy的基本用法，了解numpy.linalg包多维数组和矩阵计算函数，了解scipy.stats包里各种统计函数。
3. 掌握Pandas的基本用法，包括DataFrame和Series类及其函数等。
4. 掌握Matplotlib绘图的基本用法，包括直线、曲线、散点图、直方图等。
5. 以前接触过StatsModels尤佳。
6. 学习Python可以参考《Python核心编程》第二版，建议读第二版。本书虽然以Python 2为目标语言，但是比第三版讲的好，涉及更多Python语言的核心特征和基础知识。Python的2、3版本虽然有所区别，内核基本是一致的。<br>

<br>
7. 各Python包的学习网站如下：

   - Numpy: [官方文档](https://numpy.org/doc/stable/)
   - Scipy: [官方文档](https://docs.scipy.org/doc/scipy/reference/)
   - Pandas: [官方文档](https://pandas.pydata.org/docs/)
   - Matplotlib: [官方文档](https://matplotlib.org/stable/contents.html)
   - StatsModels: [项目主页](https://www.statsmodels.org/stable/index.html)
   - Patsy：[官方文档](https://patsy.readthedocs.io/en/latest/)
   - Scikit-Learn: [官方文档](https://scikit-learn.org/stable/)

### 3.3 Python开发环境：Jupyter
1. 先下载和安装Python语言基本开发环境，网址：https://www.python.org/downloads/ 。
本教程使用的3.7版本，因此建议下载版本>=3.7，最新版本是：3.9.1。建议下载3.7版本，因为python官方网站显示3.8和3.9的系列版本仍处于bugfix阶段，而3.7则是security阶段。而且3.7足够我们学习和工作使用了。
1. 本课程使用Jupyter作为开发环境，安装指南在这里：https://jupyter.org/install
1. 安装完毕后，在命令行运行jupyter-notebook命令即可启动Jupyter。
1. 本人习惯将该命令放在一个cmd文件，然后将命令文件拷贝到代码文件所在目录。每次运行点击该cmd文件，项目工作环境即代码文件所在目录，这样更方便。注意：要将jupyter-notebook命令文件所在文件夹替换成自己电脑上安装Python的路径。

### 3.4 R语言
1. 能读懂R代码即可。本课程中有少部分内容需要和R的计算结果对比验证。

1. R环境可先安装R语言，网址：https://www.r-project.org/

1. R和RStudio的安装过程可以参考这篇帖子：https://www.jianshu.com/p/1a0f25086e8b
